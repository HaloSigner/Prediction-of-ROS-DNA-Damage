```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Load Data

x_df = pd.read_csv('cell_painting_modz.tsv', sep = '\t')

# Separate the metadata columns

x_meta = x_df.iloc[:, :3]

# Drop the metadata columns from the original DataFrame

x_df = x_df.iloc[:, 3:]

y_all_df = pd.read_csv('cell_health_modz.tsv', sep = '\t')

y_df = y_all_df.loc[:, ['vb_ros_mean']]
```

```python
# Compute the correlation matrix

corr_matrix = x_df.corr().abs()

# remove high_corr features

to_drop = []
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if (corr_matrix.iloc[i, j] >= 0.9) and (corr_matrix.columns[j] not in to_drop):
            to_drop.append(corr_matrix.columns[i])
        
print(to_drop)

x_filt_df = x_df.drop(to_drop, axis = 1)

display(x_df)
display(x_filt_df)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor


# Define VIF Calculate function

def calculate_vif(df):
    vif_df = pd.DataFrame()
    vif_df['Feature'] = df.columns
    vif_df['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    return vif_df

# Compute VIF

vif_result = calculate_vif(x_df)

print(vif_result)
```
