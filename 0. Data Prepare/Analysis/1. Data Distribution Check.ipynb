```python
# Load Module

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Input data

x = pd.read_csv('cell_painting_modz.tsv', sep = '\t')
x
```

```python
# Load Lable data

y = pd.read_csv('cell_health_modz.tsv', sep = '\t')
y
```

```python
# Cell Health Features checking

y.columns
```

```python
# ROS value check by histogram

ros = y.loc[:, ['vb_ros_mean', 'vb_ros_back_mean']] # vb_ros_mean, vb_ros_back_mean

live_cell = y.loc[:, 'cc_all_n_objects']

dna_damage = y.loc[:, ['cc_all_n_spots_h2ax_mean', 'cc_all_n_spots_h2ax_per_nucleus_area_mean', 'cc_all_high_h2ax']]

# Setting the style for the plots to match Nature style
sns.set(style="whitegrid")

# Creating a figure with subplots for a Nature-style layout
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 18))

# Flatten axes for easy iteration
axes = axes.flatten()

def add_stats_to_legend(ax, data, color):
    median = np.median(data)
    variance = np.var(data)
    std_dev = np.std(data)
    count = len(data)  # Number of data points
    label = f'Median: {median:.2f}\nVariance: {variance:.2f}\nStd Dev: {std_dev:.2f}\nCount: {count}'
    ax.legend([label], loc='upper right', fontsize=12, frameon=True, handletextpad=1.0, borderpad = 1.0)

# Histogram for vb_ros_mean
sns.histplot(ros['vb_ros_mean'], bins=10, kde=False, color='blue', ax=axes[1], edgecolor='black')
axes[1].set_title('Histogram of vb_ros_mean', fontsize=16, weight='bold')
axes[1].set_xlabel('ROS', fontsize=14)
axes[1].set_ylabel('Frequency', fontsize=14)
add_stats_to_legend(axes[1], ros['vb_ros_mean'], 'blue')

# Histogram for vb_ros_back_mean
sns.histplot(ros['vb_ros_back_mean'], bins=10, kde=False, color='cyan', ax=axes[2], edgecolor='black')
axes[2].set_title('Histogram of vb_ros_back_mean', fontsize=16, weight='bold')
axes[2].set_xlabel('ROS Background', fontsize=14)
axes[2].set_ylabel('Frequency', fontsize=14)
add_stats_to_legend(axes[2], ros['vb_ros_back_mean'], 'cyan')

# Histogram for cc_all_n_objects
sns.histplot(live_cell, bins=10, kde=False, color='green', ax=axes[0], edgecolor='black')
axes[0].set_title('Histogram of cc_all_n_objects', fontsize=16, weight='bold')
axes[0].set_xlabel('Viability', fontsize=14)
axes[0].set_ylabel('Frequency', fontsize=14)
add_stats_to_legend(axes[0], live_cell, 'green')

# Histogram for cc_all_n_spots_h2ax_mean
sns.histplot(dna_damage['cc_all_n_spots_h2ax_mean'], bins=10, kde=False, color='red', ax=axes[5], edgecolor='black')
axes[5].set_title('Histogram of cc_all_n_spots_h2ax_mean', fontsize=16, weight='bold')
axes[5].set_xlabel('DNA Damage (Mean Spots)', fontsize=14)
axes[5].set_ylabel('Frequency', fontsize=14)
add_stats_to_legend(axes[5], dna_damage['cc_all_n_spots_h2ax_mean'], 'red')

# Histogram for cc_all_n_spots_h2ax_per_nucleus_area_mean
sns.histplot(dna_damage['cc_all_n_spots_h2ax_per_nucleus_area_mean'], bins=10, kde=False, color='purple', ax=axes[3], edgecolor='black')
axes[3].set_title('Histogram of cc_all_n_spots_h2ax_per_nucleus_area_mean', fontsize=16, weight='bold')
axes[3].set_xlabel('DNA Damage (Spots per Nucleus Area)', fontsize=14)
axes[3].set_ylabel('Frequency', fontsize=14)
add_stats_to_legend(axes[3], dna_damage['cc_all_n_spots_h2ax_per_nucleus_area_mean'], 'purple')

# Histogram for cc_all_high_h2ax
sns.histplot(dna_damage['cc_all_high_h2ax'], bins=10, kde=False, color='orange', ax=axes[4], edgecolor='black')
axes[4].set_title('Histogram of cc_all_high_h2ax', fontsize=16, weight='bold')
axes[4].set_xlabel('High DNA Damage', fontsize=14)
axes[4].set_ylabel('Frequency', fontsize=14)
add_stats_to_legend(axes[4], dna_damage['cc_all_high_h2ax'], 'orange')

# Adjusting the layout for better presentation
plt.tight_layout()

# Save the figure
plt.savefig('MODZ data distribution.png', dpi=300)

plt.show()
```
