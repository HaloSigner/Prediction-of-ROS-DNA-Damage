```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Data
total_dna = 105
brd_name = 1
dna_confirmed = 13

# 4. Donut Chart
fig, ax = plt.subplots(figsize=(12, 6))

donut_labels = ["BRD Compounds", "Non DNA Related or Cytotoxic Compounds", "DNA Related or Cytotoxic Compounds"]
donut_values = [total_dna, brd_name, dna_confirmed]
bar_colors = ['skyblue', 'orange', 'lightgreen']

# Create the donut chart
wedges, texts, autotexts = ax.pie(donut_values, labels=donut_labels, autopct='%1.1f%%', colors=bar_colors, startangle=180, counterclock=False)
plt.setp(wedges, width=0.3)

# Adjust the text size
plt.setp(texts, fontsize=10)          # Font size for labels
plt.setp(autotexts, fontsize=9)      # Font size for percentage text
ax.set_title('Predicted DNA Damaging Compounds', fontsize=14)  # Font size for the title

# Adjust legend
ax.legend(wedges, donut_labels, title="Categories", title_fontsize='13', loc="upper left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=8)

plt.savefig('DNA.png', dpi=800)
plt.tight_layout()
plt.show()
```

```python
# Data
total_dna = 105
brd_name = 1
dna_confirmed = 13


# Data for the revised donut chart
donut_labels_revised = ["Non DNA Related or Cytotoxic Compounds", "DNA Related or Cytotoxic Compounds"]
donut_values_revised = [brd_name, dna_confirmed]
bar_colors_revised = ['orange', 'lightgreen']

# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(24, 12))  # 1 row, 2 columns

# Original donut chart
wedges1, texts1, autotexts1 = axs[0].pie(
    [total_dna, brd_name, dna_confirmed], 
    labels=["BRD Compounds", "Non DNA Related or Cytotoxic Compounds", "DNA Related or Cytotoxic Compounds"], 
    autopct='%1.1f%%', 
    colors=['skyblue', 'orange', 'lightgreen'], 
    startangle=180,  # Rotate the chart by 90 degrees to the left
    counterclock=False
)
plt.setp(wedges1, width=0.3)  # Create a donut chart
axs[0].set_title('All DNA-Damaging Compounds', fontsize=14)

# Revised donut chart
wedges2, texts2, autotexts2 = axs[1].pie(
    donut_values_revised, 
    labels=donut_labels_revised, 
    autopct='%1.1f%%', 
    colors=bar_colors_revised, 
    startangle=180,  # Rotate the chart by 90 degrees to the left
    counterclock=False
)
plt.setp(wedges2, width=0.3)  # Create a donut chart
axs[1].set_title('Selected Non BRD DNA-Damaging Compounds', fontsize=14)

# Adjust legend for both charts
axs[0].legend(wedges1, ["BRD Compounds", "Non DNA Related or Cytotoxic Compounds", "DNA Related or Cytotoxic Compounds"], 
              title="Categories", title_fontsize='13', loc="upper left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=8)
axs[1].legend(wedges2, donut_labels_revised, 
              title="Categories", title_fontsize='13', loc="upper left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=8)

# Adjust layout
plt.tight_layout()
plt.savefig('Combined_DNA_Charts.png', dpi=800)
plt.show()
```
