```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import seaborn as sns
```

```python
# Load Compound List
dna_comp = pd.read_excel('Compound List(DNA Damage).xlsx')
ros_comp = pd.read_excel('Compound List(ROS Induced).xlsx')

display(dna_comp)
display(ros_comp)
```

```python
# Find intersected compound list
dna_set = set(dna_comp['CPD_NAME'])
ros_set = set(ros_comp['CPD_NAME'])


# intersection
intersection = dna_set & ros_set
print(len(intersection))

# Visualization
sns.set(style="whitegrid")
fig, ax = plt.subplots(figsize=(12, 8))
venn = venn2([dna_set, ros_set], ('DNA Damaging Compound', 'ROS Inducing Compound'))
plt.title('Venn Diagram of Predicted Compound Intersection', fontsize=14)

for text in venn.set_labels:
    text.set_fontsize(14)
for text in venn.subset_labels:
    if text: 
        text.set_fontsize(17)
        text.set_color('black')

plt.tight_layout()
plt.savefig('Intersected Compound.png', dpi = 300)
plt.show()
```

```python
# Save Intersected Compound List
chem_list = pd.read_csv('chemical_annotations.csv')


inter_chem = chem_list[chem_list['CPD_NAME'].isin(intersection)]
inter_chem.to_excel('Intersected_compound.xlsx')
```
