```python
x_df = pd.read_csv('x_df.csv', index_col = None)
x_df = x_df.iloc[:, 1:]
x_df
```

```python
# Load cell health data
y_all_df = pd.read_csv('cell_health_modz.tsv', sep = '\t')
y_df = y_all_df['cc_all_n_spots_h2ax_per_nucleus_area_mean']
```

```python
# Laod Synthetic Data
synthetic_data = pd.read_csv('synthetic_data(FastML).csv')
synthetic_data = synthetic_data.iloc[:, 1:]
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
import torch
import random

# Fix random seeds for reproducibility
def set_seed(seed_value):
    np.random.seed(seed_value)
    random.seed(seed_value)
    torch.manual_seed(seed_value)
    

# Set the seed
seed_value = 42
set_seed(seed_value)
```

```python
# Load Label Data
from collections import Counter
y_df_binarized = np.load('y_df_binarized.npy')
Counter(y_df_binarized)
```

## Data Split

```python
# train_test split
from sklearn.model_selection import train_test_split, GridSearchCV

# Assign Synthetic Data
x_df_s = synthetic_data.drop(columns='dna')
y_df_s = synthetic_data['dna']


# Data Split
X_train, X_test, y_train, y_test = train_test_split(x_df_s, y_df_s, test_size = 0.15, stratify=y_df_s, random_state=42)

# Data Flatten
y_train = np.ravel(y_train)
y_test = np.ravel(y_test)


print(f'Train Label:{Counter(y_train)}')
print(f'Train Label:{Counter(y_test)}')
```

```python
# Divied Real Data
X_train_10, X_test_10, y_train_10, y_test_10 = train_test_split(x_df, y_df_binarized, test_size = 0.90, stratify=y_df_binarized, random_state=42)

X_train_30, X_test_30, y_train_30, y_test_30 = train_test_split(x_df, y_df_binarized, test_size = 0.70, stratify=y_df_binarized, random_state=42)

X_train_50, X_test_50, y_train_50, y_test_50 = train_test_split(x_df, y_df_binarized, test_size = 0.50, stratify=y_df_binarized, random_state=42)

# Data Flatten
y_train_10 = np.ravel(y_train_10)
y_test_10 = np.ravel(y_test_10)

y_train_30 = np.ravel(y_train_30)
y_test_30 = np.ravel(y_test_30)

y_train_50 = np.ravel(y_train_50)
y_test_50 = np.ravel(y_test_50)


print('Divied 10% ----------------------------------------------------\n')
print(f'Train Label:{Counter(y_train_10)}')
print(f'Test Label:{Counter(y_test_10)}\n')

print('Divied 30% ----------------------------------------------------\n')
print(f'Train Label:{Counter(y_train_30)}')
print(f'Test Label:{Counter(y_test_30)}\n')

print('Divied 50% ----------------------------------------------------\n')
print(f'Train Label:{Counter(y_train_50)}')
print(f'Test Label:{Counter(y_test_50)}')
```

## 10% Fine Tuning
