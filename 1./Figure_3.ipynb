```python
import pandas as pd
import numpy as np
```

```python
syn_result = pd.read_excel('Synthetic Data Results.xlsx')
syn_result
```

```python
syn_result = syn_result.T
syn_result.columns = syn_result.iloc[0].tolist()
syn_result = syn_result.iloc[1:, :]
```

```python
syn_result.columns = ['Inter-row Similarity', 'Column-wise Similarity', 'Combined Similiarty']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))
ax = syn_result.plot(kind='bar', figsize=(14, 10), alpha=0.8)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',
               (p.get_x() + p.get_width() / 2., p.get_height()),
               ha='center',
               va='bottom',
               fontsize=10,
               color='black',
               rotation=0, 
               xytext=(0, 5),
               textcoords='offset points')


plt.title('Generative Model Performance Comparison', fontsize=20, fontweight='bold', color='black')
plt.xlabel('Models', fontsize=14)
plt.ylabel('Scores', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.legend(title='Score', fontsize=12, title_fontsize='13', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('Gnerative Model Performance.png', dpi=300)
plt.show()
```
