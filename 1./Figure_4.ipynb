```python
import pandas as pd
import numpy as np
```

```python
syn_res = pd.read_excel('DNA damage model result(Syn).xlsx')
syn_res.index = syn_res['Unnamed: 0'].tolist()
syn_res = syn_res.drop(columns='Unnamed: 0')
```

```python
real_res = pd.read_excel('DNA damage model results(real).xlsx')
real_res.index = real_res['Unnamed: 0'].tolist()
real_res = real_res.drop(columns='Unnamed: 0')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


ax = syn_res.plot(kind='bar', figsize=(14, 10), alpha=0.8)
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',
               (p.get_x() + p.get_width() / 2., p.get_height()),
               ha='center',
               va='bottom',
               fontsize=10,
               color='black',
               rotation=0,
               xytext=(0, 5),
               textcoords='offset points')


plt.title('Synthetic + Real Data Model Performance', fontsize=20, fontweight='bold', color='black')
plt.xlabel('Models', fontsize=14)
plt.ylabel('Scores', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.legend(title='Score', fontsize=12, title_fontsize='13', loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('Synthetic + Real Data Performance.png', dpi=300)
plt.show()
```

```python
ax = real_res.plot(kind='bar', figsize=(14, 10), alpha=0.8)
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',
               (p.get_x() + p.get_width() / 2., p.get_height()),
               ha='center',
               va='bottom',
               fontsize=10,
               color='black',
               rotation=0,
               xytext=(0, 5),
               textcoords='offset points')


plt.title('Real Data Model Performance', fontsize=20, fontweight='bold', color='black')
plt.xlabel('Models', fontsize=14)
plt.ylabel('Scores', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.legend(title='Score', fontsize=12, title_fontsize='13', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('Real Data Performance.png', dpi=300)
plt.show()
```
