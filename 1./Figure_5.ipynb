```python
import pandas as pd
```

```python
# load compound datab
dna_comp = pd.read_excel('DNA_Damage.xlsx')
dna_comp = dna_comp.iloc[:, 1:]
dna_comp
```

```python
x_val_df = pd.read_csv('x_val_dff.csv')
x_val_df = x_val_df.iloc[:, 1:]
x_val_df
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns


x_val_dff = x_val_df.iloc[:, 1:]

# PCA
pca = PCA()
x_val_pca = pca.fit_transform(x_val_dff)

# explaned variance rate
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = explained_variance_ratio.cumsum()

# visualization
plt.figure(figsize=(10, 6))
plt.plot(cumulative_variance_ratio, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by PCA Components', fontsize=16)
plt.xlabel('Number of Components', fontsize=14)
plt.ylabel('Cumulative Explained Variance', fontsize=14)
plt.grid(True)
plt.show()

# 95% variance
n_components = next(i for i, total in enumerate(cumulative_variance_ratio) if total >= 0.95) + 1
print(f"Number of components explaining 95% of the variance: {n_components}")

# PCA with the determined number of components
pca = PCA(n_components=n_components)
x_val_pca = pca.fit_transform(x_val_dff)
```

```python
# t-SNE
tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
x_val_tsne = tsne.fit_transform(x_val_pca)

# Results to dataframe
x_val_dff['tsne_1'] = x_val_tsne[:, 0] # axis 1
x_val_dff['tsne_2'] = x_val_tsne[:, 1]  # axis 2
```

```python
# Match series
match_series = x_val_df['Metadata_broad_sample'].isin(dna_comp['BROAD_ID'])
x_val_dff['cluster'] = match_series.astype(int)
```

```python
from scipy.spatial.distance import cdist

def get_closest_points(blue_points, red_points, x_val_df, dna_comp, n):
 
    
    # Calculate Euclidean
    distances_blue_to_red = cdist(blue_points.values, red_points.values, metric='euclidean')
    
    
    min_distances_blue = distances_blue_to_red.min(axis=1)
    
    
    closest_indices = min_distances_blue.argsort()[:n]


    closest_points = x_val_df.iloc[closest_indices]


    hit_list = x_val_df.iloc[closest_indices]['Metadata_broad_sample']
    
    
    cpd_names = dna_comp.loc[dna_comp['BROAD_ID'].isin(hit_list), 'CPD_NAME']

    
    return closest_points, hit_list, cpd_names



blue_points = x_val_dff[x_val_dff['cluster'] == 1][['tsne_1', 'tsne_2']]
red_points = x_val_dff[x_val_dff['cluster'] == 2][['tsne_1', 'tsne_2']]

# Find top n closest points
n = 100
closest_points, hit_list, cpd_names = get_closest_points(blue_points, red_points, x_val_df, dna_comp, n)

# Results Print
print("Hit List of Broad Sample IDs:")
print(f'Number of 100 Closest Compound: {len(hit_list)}')
print(hit_list)

print("\nHit Compound Names:")
print(f'Number of Blue Point Compound: {len(cpd_names)}')
print(cpd_names)
```

```python
closest_comp = dna_comp[dna_comp['CPD_NAME'].isin(cpd_names.tolist())]
closest_comp.to_csv('Closest_comp(100).csv')
```

```python
import umap

umap_model = umap.UMAP(n_neighbors=45, min_dist=0.1, n_components=2, random_state=42)
X_umap = umap_model.fit_transform(x_val_dff)

x_val_dff['umap_1'] = X_umap[:, 0]
x_val_dff['umap_2'] = X_umap[:, 1]
```
