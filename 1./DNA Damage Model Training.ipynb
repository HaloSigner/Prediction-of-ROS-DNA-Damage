## Module Importing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
warnings.filterwarnings(action='ignore')
```

## Real Cell Painting Data Load

```python
x_df = pd.read_csv('x_df.csv')
x_df = x_df.iloc[:, 1:]
x_df
```

## Label Data Loading

```python
y_df_binarized = np.load('y_df_binarized.npy')
```

## Data Merging for synthesize data

```python
full_df = x_df
full_df['dna'] = y_df_binarized
full_df
```

## Load Synthesize Data

```python
synthetic_data = pd.read_csv('Synthetic_data(Gaussian Copula).csv')
synthetic_data = synthetic_data.iloc[:, 1:]
```

```python
# Filter Out Class 1 Data
synthetic_data_1 = synthetic_data[synthetic_data['dna'] == 1]
synthetic_data_1
```

## Sampling Real Data (Class 0 : 30 , Class 1 : 30) for evade data leakage 

```python
df_0 = full_df[full_df['dna'] == 0].sample(n=30, random_state=42)
df_1 = full_df[full_df['dna'] == 1].sample(n=30, random_state=42)


result_df = pd.concat([df_0, df_1])


full_df = full_df.drop(result_df.index)
```

```python
# Combined Real data + Synthetic Data
combined_data = pd.concat([full_df, synthetic_data_1], axis=0)
combined_data
```

```python
from sklearn.model_selection import train_test_split
from collections import Counter

x_df_s = combined_data.drop(columns='dna')
y_df_s = combined_data['dna']
#x_df = x_df.drop(columns='dna')


#from sklearn.preprocessing import MinMaxScaler

#scaler = MinMaxScaler()

# Data Split
X_train, X_test, y_train, y_test = train_test_split(x_df_s, y_df_s, stratify=y_df_s, test_size = 0.15, random_state = 42)

# Data Scailing
#X_train = scaler.fit_transform(X_train)
#X_test = scaler.transform(X_test)
#X_val = scaler.transform(x_df)

#X_train = pd.DataFrame(X_train, columns=x_df_s.columns)
#X_test = pd.DataFrame(X_test, columns=x_df_s.columns)
#X_val = pd.DataFrame(X_val, columns=x_df.columns)

#X_train = pd.concat([X_train, synthetic_data.drop(columns=['dna'])], ignore_index=True)
#y_train = pd.concat([pd.Series(y_train), synthetic_data['dna']], ignore_index=True)

# Data Flatten
y_train = np.ravel(y_train)
y_test = np.ravel(y_test)


# Check data stratified
print(Counter(y_train))
print(Counter(y_test))
```

## SVM

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.svm import SVC
from sklearn.metrics import f1_score, classification_report, roc_auc_score


# Class weight Setting
class_weight = {0: 2, 1: 1.7}

# SVM Model Setting
svm_model = SVC(random_state=42, class_weight=class_weight)

# SVM Parmeter Setting
svm_param_grid = {
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto'],
    'kernel': ['linear', 'rbf', 'poly']
}

# Best Parameter Searching
results = {}
result = {}

# Stratified K-Fold Setting
stratified_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("Training SVM...")
grid_search = GridSearchCV(estimator=svm_model, param_grid=svm_param_grid, cv=stratified_kfold, scoring='roc_auc_ovr', verbose=1, n_jobs=-1)
grid_search.fit(X_train, y_train)

# Best Parameters & 5 Fold-cross validation
cv_results = grid_search.cv_results_

fold_f1_scores = []
for i in range(5):
    print(f'Fold {i+1} Training ................')

print(f"Best Parameters for SVM:")
print(grid_search.best_params_)

# Prediction with best estimator
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# AUC ROC Calculation
auc_roc = roc_auc_score(y_test, best_model.decision_function(X_test))

# Assessment score present
print(f"\nTest Result for SVM:")
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Generate classification report as a dictionary
report = classification_report(y_test, y_pred, output_dict=True)

# Store result 
result['SVM'] = {
    'f1_score': report['weighted avg']['f1-score'],
    'accuracy': report['accuracy']
}

# Final Results Reporting
print("\nSVM Test Results:")
print(f"SVM F1-score : {result['SVM']['f1_score']: .4f}, AUC ROC : {auc_roc:.4f}, Accuracy : {result['SVM']['accuracy']:.4f}")
```

```python
# Validation with sampled Real Data
y_pred = grid_search.best_estimator_.predict(result_df.drop(columns = 'dna'))

print(classification_report(result_df['dna'], y_pred))
```

## Result Visualization

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc

conf_matrix = confusion_matrix(result_df['dna'], y_pred)

plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig('Confusion_mat.png', dpi=300)
plt.show()

from sklearn.metrics import roc_curve, auc

# Decision_Funciton score calculation
y_scores = grid_search.best_estimator_.decision_function(result_df.drop(columns='dna'))

# ROC Curve & AUC calculation
fpr, tpr, _ = roc_curve(result_df['dna'], y_scores)
roc_auc = auc(fpr, tpr)

# ROC Curve Visualization
plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.savefig('ROC_Curve.png', dpi=300)
plt.show()
```

## LightGBM

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, roc_auc_score

# LightGBM Model Setting
LGBM_model = LGBMClassifier(random_state=42, scale_pos_weight=3) 

# LightGBM Parameter Setting
LGBM_param_grid = {
    'learning_rate': [0.01, 0.1, 0.3],
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7]
}

# Stratified K-Fold Setting
stratified_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("Training LightGBM...")
grid_search = GridSearchCV(estimator=LGBM_model, param_grid=LGBM_param_grid, cv=stratified_kfold, scoring='roc_auc_ovr', verbose=1, n_jobs=-1)
grid_search.fit(X_train, y_train)

# Best Parameters & 5 Fold-cross validation
print(f"Best Parameters for LightGBM:")
print(grid_search.best_params_)

# Prediction with best estimator
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# AUC ROC Calculation
auc_roc = roc_auc_score(y_test, best_model.predict_proba(X_test)[:, 1])

# Assessment score present
print(f"\nTest Result for LightGBM:")
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Generate classification report as a dictionary
report = classification_report(y_test, y_pred, output_dict=True)

# Store result
result = {
    'LightGBM': {
        'f1_score': report['weighted avg']['f1-score'],
        'accuracy': report['accuracy']
    }
}

# Final Results Reporting
print("\nLightGBM Test Results:")
print(f"LightGBM F1-score : {result['LightGBM']['f1_score']: .4f}, AUC ROC : {auc_roc:.4f}, Accuracy : {result['LightGBM']['accuracy']:.4f}")
```

## CPG-0012 Prediction

```python
# Load cpg0012 data
cpg = pd.read_csv('x_val_df.csv')
cpg = cpg.iloc[:, 1:]
cpg
```

```python
# Load cpg0012 metadata
cpg_meta = pd.read_csv('x_val_dff.csv')
cpg_meta = cpg_meta.iloc[:, 1:]
cpg_meta
```

```python
# Chemical annotation 
chem_dat = pd.read_csv('chemical_annotations.csv')
chem_dat
```

```python
# Predict with SVM and Counting
y_pred_1 = grid_search.best_estimator_.predict(cpg)
Counter(y_pred_1)
```

```python
# Filtering Class 1 with index
indices_with_1 = [i for i, pred in enumerate(y_pred_1) if pred == 1]

print(len(indices_with_1))
```

```python
# Initialize a list to store compound names
comp_name = []

# Iterate over predicted values (assuming y_pred_1 has the right indices)
for idx in range(len(indices_with_1)):
    # Get the corresponding 'Metadata_broad_sample' value from x_val_dff
    value = cpg_meta.iloc[idx]['Metadata_broad_sample']  # Assuming idx matches rows in x_val_dff
    # Look up the compound name in chem_dat by matching BROAD_ID
    name = chem_dat[chem_dat['BROAD_ID'] == value]['CPD_NAME'].tolist()
    comp_name.append(name)

# Remove DMSO (i.e., empty lists or non-chemical names)
comp_name = [v for v in comp_name if v]

# Print the number of valid compound names
print(len(comp_name))

# Display the list of compound names (removing the unnecessary print of display())
display(comp_name)
```

```python
# Filter out only class 1 
filtered_comp_name = [
    name for name in comp_name 
    if all(isinstance(n, str) and not n.startswith('BRD') for n in name)
]

# Flatten the list
filtered_comp_name = [item for sublist in filtered_comp_name for item in sublist]

print(len(filtered_comp_name))
display(filtered_comp_name)
```

```python
# Save Non-BRD Compound only
dna_damage = chem_dat[chem_dat['CPD_NAME'].isin(filtered_comp_name)]
dna_damage.to_csv('DNA_Damage.csv')
```
